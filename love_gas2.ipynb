```python
# 

from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt

# 
from sklearn.linear_model import LinearRegression
mod = LinearRegression()

X = df_train.loc[:, ['weight']].values
y = df_train.loc[:, 'mpg'].values

# 
df_train.loc[:, 'mpg'].plot.hist()
plt.show()

poly_quad  = PolynomialFeatures(degree = 2)# 2
poly_cubic = PolynomialFeatures(degree = 3)# 3

# 
X_quad = poly_quad.fit_transform(X)
X_cubic = poly_cubic.fit_transform(X)

X_fit = np.arange(X.min(), X.max(), 1)[:, np.newaxis]

# R^2
mod_lin = mod.fit(X, y)
y_lin_fit = mod_lin.predict(X_fit)
r2_lin = mod.score(X, y)

# 2R^2
mod_quad = mod.fit(X_quad, y)
y_quad_fit = mod_quad.predict(poly_quad.fit_transform(X_fit))
r2_quad = mod.score(X_quad, y)

# 3R^2
mod_cubic = mod.fit(X_cubic, y)
y_cubic_fit = mod_cubic.predict(poly_cubic.fit_transform(X_fit))
r2_cubic = mod.score(X_cubic, y)
```

```python
plt.scatter(X, y, label= 'training data', color= 'gray')
plt.plot(X_fit, y_lin_fit, label= 'linear', color = 'black', linestyle = ':')
plt.plot(X_fit, y_quad_fit, label = 'quadratic', color= 'orange', lw = 2, linestyle = '-')
plt.plot(X_fit, y_cubic_fit,  label = 'cubic',  color = 'red', linestyle = '--')
plt.legend(loc = 'upper right')

plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```
